## **1. Supervised Learning (Predicting with Labels)**

### **A. Regression (Predict Numbers)**
 **Example:** Predict ice cream sales üç¶ based on temperature.

```python
# Import libraries
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Training data: [Temperature (¬∞C)] ‚Üí [Sales]
X = np.array([[20], [25], [30], [35]])  # Features
y = np.array([50, 80, 120, 150])       # Labels

# Train the model
model = LinearRegression()
model.fit(X, y)

# Predict for 28¬∞C
prediction = model.predict([[28]])
print(f"Predicted sales: {prediction[0]:.0f} ice creams")

# Plot
plt.scatter(X, y, color='blue', label='Actual Sales')
plt.plot(X, model.predict(X), color='red', label='Predicted Sales')
plt.title("Ice Cream Sales vs Temperature")
plt.xlabel("Temperature (¬∞C)")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

### **B. Classification (Predict Categories)**
 **Example 1:** Binary (Diabetes Yes/No) üè•

```python
from sklearn.tree import DecisionTreeClassifier

# Training data: [Age, Blood Sugar] ‚Üí [Diabetic?]
X = np.array([[30, 120], [40, 150], [50, 200], [60, 250]])
y = np.array([0, 0, 1, 1])  # 0=Healthy, 1=Diabetic

# Train the model
model = DecisionTreeClassifier()
model.fit(X, y)

# Predict for a new patient
prediction = model.predict([[45, 180]])
print("Diabetic?", "Yes" if prediction[0] == 1 else "No")
```

**Example 2:** Multiclass (Penguin Species) üêß

```python
from sklearn.svm import SVC

# Training data: [Flipper Length (mm), Bill Length (mm)] ‚Üí [Species]
X = np.array([[180, 40], [190, 45], [210, 50], [220, 55]])
y = np.array([0, 0, 1, 2])  # 0=Adelie, 1=Gentoo, 2=Chinstrap

# Train the model
model = SVC()
model.fit(X, y)

# Predict for a new penguin
prediction = model.predict([[200, 48]])
species = ["Adelie", "Gentoo", "Chinstrap"]
print("Species:", species[prediction[0]])
```

**2. Unsupervised Learning (Finding Patterns)**
 **Example:** Clustering Customers üõí

```python
from sklearn.cluster import KMeans

# Customer data: [Age, Annual Spending ($1000s)]
X = np.array([[25, 20], [30, 25], [45, 50], [50, 60]])

# Train the model (2 clusters)
model = KMeans(n_clusters=2)
model.fit(X)

# Predict clusters
labels = model.predict(X)
print("Cluster groups:", labels)  # e.g., [0, 0, 1, 1]

# Plot
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow')
plt.title("Customer Clusters")
plt.xlabel("Age")
plt.ylabel("Annual Spending ($1000s)")
plt.show()
```
